```python
import cv2
import mediapipe as mp
import numpy as np
```

```python

cap = cv2.VideoCapture(0)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        print("Ignorando o frame vazio da camera.")
        continue

    cv2.imshow('Camera', frame)

    if cv2.waitKey(10) & 0xFF == ord('c'):  # Pressione 'c' para capturar
        break
cap.release()
cv2.destroyAllWindows()
```
